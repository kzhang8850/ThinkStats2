```python
%matplotlib inline

import thinkstats2, thinkplot
import pandas as pd
import numpy as np
```

```python
def clean_dataframe(data):
    """
    Takes entire Pandas dataframe and cleans out 66, 77, 88, and 99, which are all bad responses, by replacing them with
    NaNs. Can then use dropna() to fully clean out data later.
    
    Input: Pandas dataframe
    Output: dataframe that has been cleaned out with NaNs
    """
    for series in data.columns:
        data[series].replace([66, 77, 88, 99], np.nan, inplace=True)
    return data
```

```python
def make_and_plot_pmf(database, category, label):
    """
    Takes a category and makes a pmf. Then also plots the pmf using hist. Returns the pmf.
    
    Input: Category string
    Output: pmf
    """
    series = database[category].dropna()
    series_pmf = thinkstats2.Pmf(series, label=label)
    thinkplot.Hist(series_pmf)
    thinkplot.Config(xlabel=label, ylabel="pmf")
    
    return series_pmf
```

```python
ess5 = pd.read_stata('../../ESS5e03_3.dta',convert_categoricals=False,convert_missing=False)
ess5 = clean_dataframe(ess5)

# Look into all the politics stuff(left vs right and informed, liberal policies, trust, as well as happy and total income

```
